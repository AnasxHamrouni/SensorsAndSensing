### **Electronic sensor**

Есть два типа матриц: CMOS (Complementary Metal-Oxide Semiconductor) и CCD (Charge-Coupled Device).
На раннем этапе развития матриц технология CCD являлась эффективней CMOS, благодаря:
* Низкий уровень шумов;
* Высокий коэффициент заполнения пикселов;
* Достаточно высокая эффективность;
* Высокая светочувствительность.

Список преимуществ CMOS-технологии:
- во-первых, низкое энергопотребление в статическом состоянии по сравнению с CCD
- во-вторых, CMOS сразу «выдаёт» цифровой сигнал, который не требует дополнительного преобразования, в отличие от CCD, которое является фактически аналоговым устройством
- в-третьих, дешевизна производства, особенно при больших размерах матриц

There are two types of matrices: CMOS (Complementary Metal-Oxide Semiconductor) and CCD (Charge-Coupled Device).
In the early stage of matrix development, CCD technology was more efficient than CMOS because of:
* Low noise;
* High pixel fill factor;
* Sufficiently high efficiency;
* High light sensitivity.

The list of advantages of CMOS technology:
- firstly, low power consumption in static state compared to CCD
- Secondly, CMOS immediately "produces" a digital signal, which does not require additional conversion, unlike CCD, which is actually an analog device
- thirdly, low cost of production, especially for large matrix sizes

### CCD matrix

image.png

CCD представляет собой устройство, в котором заряд от пикселей преобразуется в электрический потенциал, этот потенциал усиливается уже за пределами светочувствительных сенсоров. На первом этапе в CCD выводится изображение в аналоговом виде. А уже на втором этапе происходит оцифровка полученной картинки с помощью усилителя в аналого-цифровом преобразователе (АЦП). Таким образом, сам прибор состоит из двух частей —матрицы и преобразователя. То есть, CCD — это типичный прибор с зарядовой связью. Однако, такое устройство не дает матрице экспонировать следующий кадр, до тех пор, пока она полностью не считает данные с предыдущего. В результате этот факт приводит к необходимости использования глобального затвора (global shutter).

A CCD is a device in which the charge from the pixels is converted into an electrical potential, this potential is amplified already outside the light-sensitive sensors. At the first stage, the CCD outputs the image in analog form. And in the second stage, the resulting image is digitized using an amplifier in the analog-to-digital converter (ADC). Thus, the device itself consists of two parts - the matrix and the converter. That is, a CCD is a typical charge-coupled device. However, such a device prevents the matrix from exposing the next frame until it fully reads the data from the previous frame. This fact results in the need for a global shutter.

### CMOS matrix


image.png

CMOS включает в сенсор усилитель и АЦП. Такая процедура формирования изображения позволяет достичь гораздо большей скорости чтения. Если говорить проще, то в CMOS сенсоре цифровые усилители используются в каждом отдельном пикселе. На первый взгляд, отдельное считывание каждого пикселя должно занимать больше времени, но в CMOS сенсорах процессы считывания происходят параллельно. С помощью чего достигается высокая пропускная способность, которая выше чем у CCD. На практике в CMOS применяется плавающий затвор (Rolling Shutter), это означает, что данные считываются построчно сверху вниз.

CMOS incorporates an amplifier and ADC into the sensor. This imaging procedure allows for much faster read speeds. In simple terms, a CMOS sensor uses digital amplifiers in each individual pixel. At first glance, it should take longer to read each pixel separately, but in CMOS sensors the reading processes occur in parallel. This results in a high throughput, which is higher than that of a CCD. In practice, CMOS uses a Rolling Shutter, which means that the data is read line by line from top to bottom.

$\pi$ - image plane as the plane that contains the sensing array

The axes $x_c$ and $y_c$ - form a basis for the image plane, and are typically taken to be parallel to the horizontal and vertical axes (respectively) of the image

The axis $z_c$ is perpendicular to the image plane and aligned with the optic axis of the lens (i.e., it passes through the focal center of the lens)

Center of projection is the origin of the camera frame is located at a distance $\lambda$ behind the image plane.

Coordinates on the image plane: $(u, v, \lambda)$

Let $P$ be a point in the world with coordinates $(x, y, z)$ (relative to the camera frame). Let $p$ denote the projection of $P$ onto the image plane with coordinates $(u, v, \lambda)$. Then:
$$
kx = u\\
ky=v\\
kz=\lambda
$$
This gives $k=\frac{\lambda}{z}$ then $u=\lambda\frac{x}{z}$ and $v=\lambda\frac{y}{z}$. These are the well-known equations for **perspective projection**.

Principal point is $(o_r, o_c)$.
Indices into the pixel array of pixels is $(r, c)$
Horizontal and vertical dimensions of pixel $(s_x, s_y)$
then:
$$
-\frac{u}{s_x}=(r-o_r) \quad -\frac{v}{s_y}=(c-o_c)
$$
then:
$$
u = -s_x(r-o_r) \quad v = -s_y(c-o_c)
$$

### Distortions
Distortion is an image distortion caused by a discrepancy between the actual geometry of objects in a scene and their representation in a photograph. Distortion occurs due to the physical characteristics of the camera and lens, which can cause deformations in the angles, lengths, and shapes of objects in the image.

**Radial distortion**

Radial distortion, or lens curvature: Objects at the edges of the frame may be curved or stretched. It can be barrel-shaped (objects shrink toward the center of the frame) or pillow-shaped (objects are pulled away from the center of the frame).

**Как учитывать радиальную дисторсию?**

Нам нужно учитывать $m$ точек на $n$ изображениях и складывать все уравнения, чтобы получить $2mn$ уравнений, или $Dk = d$, где $k = [k1, k2]^T$ в матричной форме.

Линейное решение методом наименьших квадратов выглядит так:
$$
k = (D^TD)^{-1}D^Td
$$
Теперь у нас функция потерь представляется следующим образом:
$$
\text{reprojection error} = \sum_{i=1}^n\sum_{j=1}^m   \|p_{ij} -  p_{\text{proj}}(K,k_1,k_2,R_i, t_i, P_{w}^j)\|^2
$$

### **Method of connecting components**
A graph, containing vertices and connecting edges, is constructed from relevant input data. The vertices contain information required by the comparison heuristic, while the edges indicate connected 'neighbors'. An algorithm traverses the graph, labeling the vertices based on the connectivity and relative values of their neighbors. Connectivity is determined by the medium; image graphs, for example, can be 4-connected neighborhood or 8-connected neighborhood.

Following the labeling stage, the graph may be partitioned into subsets, after which the original information can be recovered and processed .
